```python
#Problem 2
#Question 1
df_m = pd.read_csv("medical_students_dataset.csv")
df_m.describe()
df_m.head(3000)
```

```python
#Question 2
df_m.info()
df_m.isnull().sum()
```

```python
#Question 3
df_m['Student ID'] = df_m['Student ID'].fillna("Unknown")
# I used unkown here as IDs are rarely used for analysis or calculations. So keepeing it as NaN or unknown can work.


df_m['Age'] = df_m['Age'].fillna(df_m['Age'].median())
#I used median value to fill the NaN values in age as the data could have been skewed with different age groups. So, filling with median is the best option.

numeric_col = ['Height', 'Weight','BMI', 'Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol']
for col in numeric_col:
  df_m[col] = df_m[col].fillna(df_m[col].median())

#All these numeric values could have outiers and median is roubust to outliers. So, again filling the NaN with median values is the best option.

cat_col = ['Gender', 'Blood Type', 'Smoking', 'Diabetes']
for col in cat_col:
  df_m[col] = df_m[col].fillna(df_m[col].mode()[0])

#Here gender, blood type, smoking and diabetes are in categorical value. So, it is better to use mode to fill the NaN with the highest occuring value.
```

```python
#Question 4
df_m= df_m.drop_duplicates(keep='first')
```

```python
#3.2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_t = pd.read_csv("Titanic-Dataset.csv")
df_t.head()
```

```python
#Problem 1
subset_col = ['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']
df_subset = df_t[subset_col].copy()

df_first_class = df_subset[df_subset['Pclass']==1]

fare_mean = df_first_class['Fare'].mean()
fare_median = df_first_class['Fare'].median()
fare_max = df_first_class['Fare'].max()
fare_min = df_first_class['Fare'].min()

print("Fare Mean: ", fare_mean)
print("Fare Median: ", fare_median)
print("Maximum fare: ", fare_max)
print("Minimum fare: ", fare_min)
```

```python
#Problem 2
df_subset['Age'].isnull().sum()
df_subset = df_subset.dropna(subset=['Age'])
```

```python
#Problem 3
#Question 1
one_hot_encoding = pd.get_dummies(df_t['Embarked'], prefix='Embarked').astype(int)

#Question 2
df_t = pd.concat([df_t, one_hot_encoding], axis=1)

#Question 3
df_t = df_t.drop('Embarked', axis=1)
```

```python
#Question 4
df_t.head()
```

```python
#Problem 4
survival = df_t.groupby('Sex')['Survived'].mean()
print(survival)

survival.plot(kind ='bar', color = ['red', 'yellow'])
plt.title('Survival Rate by Gender')
plt.ylabel('Survived')
plt.ylim(0,1)
plt.show()
```

```python
emb_cols = ['Embarked_C', 'Embarked_Q', 'Embarked_S']

df_t = df_t.copy()
df_t['Embarked'] = df_t[emb_cols].idxmax(axis=1).str.replace('Embarked_', '', regex=False)

# If there are rows where all are 0 (rare), mark as Unknown
df_t.loc[df_t[emb_cols].sum(axis=1) == 0, 'Embarked'] = 'Unknown'

# Group + mean survival
survival_by_sex_embarked = (
    df_t.groupby(['Sex', 'Embarked'])['Survived']
        .mean()
        .reset_index()
)
print(survival_by_sex_embarked)

# Pivot for plotting
pivot_df_t = survival_by_sex_embarked.pivot(index='Sex', columns='Embarked', values='Survived').fillna(0)

# Ensure order C, S, Q if they exist
order = [c for c in ['C', 'S', 'Q'] if c in pivot_df_t.columns]
pivot_df_t = pivot_df_t[order]

x = np.arange(len(pivot_df_t.index))
width = 0.25

plt.figure(figsize=(8, 5))

# Plot each embarked category side-by-side
for i, col in enumerate(pivot_df_t.columns):
    plt.bar(x + (i - (len(pivot_df_t.columns)-1)/2) * width, pivot_df_t[col], width=width, label=col)

plt.xticks(x, pivot_df_t.index)
plt.title('Survival Rate by Gender and Port of Embarkation')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.legend(title='Embarked')
plt.show()
```
