```python
#Define a train-test split
def train_test_split(X, Y, test_size=0.3, random_seed=42):
  """
  Splits the dataset into train and test sets.
  Arguments:
  X: np.ndarray
     Feature Matrix
  Y: np.ndarray
     Test array
  test_size: float
     Proportion of the dataset to include in the test split.(0 < test_size < 1)
  random_seed: int
     Seed for reporducibility
  Returns:
  X_train, X_test, Y_train, Y_test: np.ndarray
     Training and testing splits of features and target.
  """
  np.random.seed(random_seed)
  indices = np.arange(X.shape[0])
  np.random.shuffle(indices)
  test_split_size = int(len(X) * test_size)
  test_indices = indices[:test_split_size]
  train_indices = indices[test_split_size:]
  X_train, X_test = X[train_indices], X[test_indices]
  Y_train, Y_test = Y[train_indices], Y[test_indices]
  return X_train, X_test, Y_train, Y_test

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)
#Output shapes to verify
print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of Y_train: ", Y_train.shape)
print("Shape of Y_test: ", Y_test.shape)
```

```python
#Define the cost function
def cost_function(X, Y, W):
  """
  This function calculates the Mean Square Error
  X: Feature Matrix
  Y: Target Matrix
  W: Weight Matrix
  Output:
  cost: Accumulated Mean Square Error.
  """
  X = np.array(X, dtype=float)
  Y = np.array(Y, dtype=float).reshape(-1, 1)
  W = np.array(W, dtype=float).reshape(-1, 1)

  n = len(Y)
  Y_predicted = X @ W
  error = Y_predicted - Y
  MSE = (1 / (n)) * np.sum(error ** 2)
  return MSE
```

```python
def rmse(Y, Y_prediction):
  """
  This Function calculates the Root Mean Squres.
  Input Arguments:
  Y: Array of actual(Target) Dependent Varaibles.
  Y_pred: Array of predeicted Dependent Varaibles.
  Output Arguments:
  rmse: Root Mean Square.
  """
  n = len(Y)
  Y = np.array(Y, dtype=float).reshape(-1, 1)
  Y_prediction = np.array(Y_prediction, dtype=float).reshape(-1, 1)
  rmse = np.sqrt(np.sum((Y - Y_prediction) ** 2)/n)
  return rmse
```

```python
def r2(Y, Y_prediction):
  """
  This function calculates the R squared error.

  Input Arguments:
  Y: Array of actual target Dependent variables
  Y_pred: Array of predicted dependent variables

  Ouput:
  rsqaured: R squared error.
  """

  mean_y = np.mean(Y)
  Y = np.array(Y, dtype=float).reshape(-1, 1)
  Y_prediction = np.array(Y_prediction, dtype=float).reshape(-1, 1)
  ss_res = np.sum((Y - Y_prediction) ** 2)
  ss_tot = np.sum((Y - mean_y) ** 2)

  r2 = 1 - ss_res/ss_tot

  return r2
```
